```python
# Exercise 4 : Data Visualisation
# Instructions
# Using the dataset: sales_data.csv

# Calculate the total quantity of products sold by the company.
# Identify the category that had the highest revenue and calculate how much revenue it generated.
# Determine the average revenue per sale in the dataset.
# Create a bar chart to visualize the total revenue generated in each quarter of the year (Q1, Q2, Q3, Q4).

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sales_data.csv")

# 1) Total quantity sold
total_qty = df['quantity'].sum()

# 2) Category with highest revenue
rev_by_cat = df.groupby('category')['revenue'].sum()
top_category = rev_by_cat.idxmax()
top_revenue = rev_by_cat.max()

# 3) Average revenue per sale (row)
avg_rev_per_sale = df['revenue'].mean()

# 4) Quarter from date (best way)
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df = df.dropna(subset=['date'])
df['Quarter'] = 'Q' + df['date'].dt.quarter.astype(str)

# 5) Total revenue per quarter + bar chart
rev_by_q = df.groupby('Quarter')['revenue'].sum().reindex(['Q1','Q2','Q3','Q4'])

plt.figure(figsize=(8, 5))
plt.bar(rev_by_q.index, rev_by_q.values)
plt.title("Total Revenue by Quarter")
plt.xlabel("Quarter")
plt.ylabel("Revenue ($)")
plt.show()

print("Total quantity sold:", total_qty)
print("Top category:", top_category, "Revenue:", top_revenue)
print("Average revenue per sale:", avg_rev_per_sale)
```

```python
# Exercise 5 : Data Visualisation Using MatPlotLib
# Instructions
# Create a simple line plot using Matplotlib. Plot the function y = x^2 for x values ranging from -10 to 10. Add a title and label the x and y axes.
# Generate a bar chart to display the sales data of four different products (A, B, C, D) in a week. The sales values are 15, 30, 45, and 20 respectively. Label the chart appropriately.
# Create a pie chart representing the following data about favorite fruits of a group of people: Apples (40%), Bananas (30%), Cherries (20%), and Dates (10%). Customize the chart with different colors for each fruit and add a legend.


# Create x values
x = np.arange(-10, 11)

# Compute y = x^2
y = x**2

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, y)

plt.title("Plot of y = x")
plt.xlabel("x values")
plt.ylabel("y values")

plt.show()



products = ['A', 'B', 'C', 'D']
sales = [15, 30, 45, 20]

plt.figure(figsize=(8, 5))
plt.bar(products, sales)

plt.title("Weekly Product Sales")
plt.xlabel("Product")
plt.ylabel("Sales Quantity")

plt.show()




fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [40, 30, 20, 10]
colors = ['red', 'yellow', 'pink', 'brown']

plt.figure(figsize=(6, 6))
plt.pie(sizes, colors=colors, autopct='%1.1f%%')

plt.title("Favorite Fruits Distribution")
plt.legend(fruits, loc="best")

plt.show()

```
