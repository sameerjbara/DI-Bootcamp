```python
# ============================================
# RUN THIS CELL FIRST - All imports for today
# ============================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Display settings
plt.style.use('seaborn-v0_8-whitegrid')
%matplotlib inline
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 20)

# Set default figure size
plt.rcParams['figure.figsize'] = (10, 6)

print("All libraries loaded! Ready for visualization.")
```

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(data=df, x='Social support', y='Score')
plt.title("Social support vs Score", fontsize=16, fontweight='bold')
plt.xlabel("Social support", fontsize=12)
plt.ylabel("Score", fontsize=12)
plt.show()
```

```python
region_data = df.groupby('Country or region')[['GDP per capita',
                                         'Healthy life expectancy']].mean()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Create figure with 2 subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# --------------------
# Subplot 1  GDP per capita
# --------------------
axes[0].bar(df["Country or region"], df["GDP per capita"])
axes[0].set_title("GDP per Capita (2019)")
axes[0].set_xlabel("Country")
axes[0].set_ylabel("GDP per capita")
axes[0].tick_params(axis='x', rotation=90)

# --------------------
# Subplot 2  Healthy Life Expectancy
# --------------------
axes[1].plot(df["Country or region"], df["Healthy life expectancy"], marker='o')
axes[1].set_title("Healthy Life Expectancy (2019)")
axes[1].set_xlabel("Country")
axes[1].set_ylabel("Healthy life expectancy")
axes[1].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Sort by GDP for cleaner visualization
df_sorted = df.sort_values("GDP per capita", ascending=False)

plt.figure(figsize=(14, 6))

# Bar plot (GDP)
plt.bar(df_sorted["Country or region"],
        df_sorted["GDP per capita"])

# Line plot (Healthy life expectancy)  different color
plt.plot(df_sorted["Country or region"],
         df_sorted["Healthy life expectancy"],
         marker='o',
         color='red')

plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("Value")
plt.title("GDP per Capita vs Healthy Life Expectancy (2019)")
plt.tight_layout()
plt.show()
```
