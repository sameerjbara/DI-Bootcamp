```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://github.com/topics"
response = requests.get(url)

print(f"Status code:  {response.status_code}")
print()


print(response.text[:100])



```

```python
with open("webpage.html", "w", encoding=response.encoding) as file:
        file.write(response.text)
```

```python

# Open the saved HTML file
with open("webpage.html", "r", encoding="utf-8") as file:
    html_content = file.read()

# Parse HTML
soup = BeautifulSoup(html_content, "html.parser")

# Example: print page title
print("Page title:", soup.title.text)
```

```python
name=soup.find('p',class_='f3 lh-condensed text-center Link--primary mb-0 mt-1').get_text().strip()

desc=soup.find('p',class_='f5 color-fg-muted text-center mb-0 mt-1').get_text().strip()
```

```python
print(f'{name} - {desc}')
```

```python
print(f'len of name: {len(name)}')
print(f'len of desc: {len(desc)}')
```

```python
data = {
    "name": [name],
    "desc": [desc]
}

df = pd.DataFrame(data)

print(df)
```
