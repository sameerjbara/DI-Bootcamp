```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# 1. Load data

df = pd.read_csv(
    "ex2data1.txt",
    header=None,
    names=["Exam1", "Exam2", "Admitted"]
)
```

```python
# 2. Visualize data

admitted = df[df["Admitted"] == 1]
not_admitted = df[df["Admitted"] == 0]

plt.scatter(admitted["Exam1"], admitted["Exam2"], label="Admitted", marker="+")
plt.scatter(not_admitted["Exam1"], not_admitted["Exam2"], label="Not admitted", marker="o")

plt.xlabel("Exam 1 Score")
plt.ylabel("Exam 2 Score")
plt.title("Student Admission Data")
plt.legend()
plt.show()
```

```python
# 3. Prepare X and y

X = df[["Exam1", "Exam2"]]
y = df["Admitted"]
```

```python
# 4. Train logistic regression model

model = LogisticRegression()
model.fit(X, y)
```

```python
# 5. Make predictions
y_pred = model.predict(X)
```

```python
# 6. Accuracy
accuracy = accuracy_score(y, y_pred)
print("Accuracy:", accuracy)

# 7. Model parameters
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```
