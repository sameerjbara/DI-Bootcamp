## Exercise 2: Multi-Part Prompt for Quiz Generation
**Scenario**: Create an educational prompt for 7th graders about volcanic eruptions.

### Tasks
- Ask for: 2-bullet summary + 3 MCQs (1 correct, 2 distractors).
- Require friendly tone, simple vocab (11?13 y/o), and Google Slides-ready output.
- Explain 3 ways your prompt improves on: `Make a quiz for kids about this article.`

```python
# TODO: write the full multi-part prompt
quiz_prompt = """
You are an educational content creator for middle school students.

Read the article about volcanic eruptions and create learning materials for 7th-grade students (ages 11–13).

Requirements:

1. Summary
   - Provide a 2-bullet summary of the article.
   - Use simple vocabulary and short sentences.
   - Focus on the most important facts about volcanic eruptions.

2. Multiple-Choice Quiz
   - Create 3 multiple-choice questions based on the article.
   - Each question must have:
     - 1 correct answer
     - 2 plausible distractors
   - Clearly label the correct answer.

3. Style Guidelines
   - Use a friendly and encouraging tone.
   - Avoid technical jargon unless it is explained.
   - Keep all content appropriate for 11–13-year-old students.

4. Output Format (Google Slides Ready)
   Slide 1: Title + 2-bullet summary
   Slide 2: Question 1
   Slide 3: Question 2
   Slide 4: Question 3
   Slide 5: Answer Key

Format the content so it can be copied directly into Google Slides.
"""
quiz_prompt
```

```python
# TODO: explain three improvements over the vague prompt
improvements = [
    "Specifies the target audience (7th graders ages 11–13), ensuring the content matches their reading level.",
    "Provides clear deliverables (2-bullet summary and 3 MCQs with one correct answer and two distractors).",
    "Defines the required tone and output structure, making the result Google Slides-ready and easy to use."
]
improvements
```

## Exercise 4: Match Prompt to Purpose
Pick a style (Exploratory, Structured, Conversational, Functional) to build a customer-support chatbot prompt.

### Tasks
- Write the full prompt in the chosen style.
- Identify 2 features that prove it matches the style.
- Justify why this style fits better than the others.

```python
# TODO: choose a style and write the prompt
chosen_style = 'Structured'

prompt_template = """
You are a customer-support chatbot for a software company.

Your responsibilities:
1. Greet customers politely.
2. Identify the customer's issue by asking clarifying questions when needed.
3. Provide accurate, step-by-step solutions.
4. If the issue cannot be resolved, escalate it to a human support agent.
5. Remain professional, patient, and helpful at all times.

Response Format:
- Issue Summary:
- Recommended Solution:
- Next Steps:

Rules:
- Use clear and concise language.
- Do not make up information.
- If information is missing, ask follow-up questions.
- Keep responses under 150 words whenever possible.
- Thank the customer before ending the conversation.
"""

key_style_features = [
    'Uses clearly defined sections, rules, and responsibilities.',
    'Provides a fixed response structure (Issue Summary, Recommended Solution, Next Steps).'
]

justification = """
A Structured style is the best choice for a customer-support chatbot because support interactions require consistency, accuracy, and predictable responses. The defined rules and format help ensure customers receive clear guidance every time. Conversational style may be friendly but can become inconsistent, Exploratory style is better for brainstorming and discovery, and Functional style focuses on completing specific tasks rather than managing full support interactions. Structured prompts provide the reliability and clarity needed for customer service.
"""
```
