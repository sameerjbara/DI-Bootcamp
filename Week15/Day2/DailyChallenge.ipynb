```python
from statsmodels.stats.power import NormalIndPower

analysis = NormalIndPower()
n = analysis.solve_power(effect_size=0.2, alpha=0.05, power=0.8,
                         ratio=1, alternative='two-sided')
print(f"{n:.0f} per group")
```

```python
for es in [0.1, 0.2, 0.3, 0.4]:
    n = analysis.solve_power(effect_size=es, alpha=0.05, power=0.8,
                             ratio=1, alternative='two-sided')
    print(f"Effect size {es}: {n:.0f} per group")
```

Sample size shrinks fast as effect size grows — it's inverse-square (n ∝ 1/effect²). Halving the effect quadruples the sample needed. Big changes are easy to spot in small samples; tiny changes hide in the noise and need lots of data to confirm.

```python
# Effect size and sample size must be matched. Big improvements show up in small samples; tiny improvements hide in the noise and need much larger samples to detect.
# Too few customers risks missing a real improvement or being fooled by a lucky week. Too many wastes time and traffic. Matching sample size to the expected effect gives just enough confidence to act on the result without overspending.
```
