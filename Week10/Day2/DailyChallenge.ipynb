```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mat=np.random.randint(-5, 35, 120).reshape(10,12)

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
cities = [
    "New York",
    "London",
    "Tokyo",
    "Paris",
    "Berlin",
    "Toronto",
    "Sydney",
    "Dubai",
    "Rome",
    "Barcelona"
]

df = pd.DataFrame(mat,
                  columns=months,
                  index=cities)


```

```python
df.head()
```

```python
annual_avg  = df.mean(axis=1)
```

```python
idxMax=annual_avg.idxmax()
print(idxMax)
idxMin=annual_avg.idxmin()
print(idxMin)
```

```python
plt.figure(figsize=(14, 8))

for city in df.index:
    plt.plot(months, df.loc[city], label=city)

plt.title("Monthly Temperature Trends for Each City", fontsize=18)
plt.xlabel("Month", fontsize=14)
plt.ylabel("Temperature (Â°C)", fontsize=14)
plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(10, 6))
annual_avg.plot(kind='bar')

plt.title("Annual Average Temperature by City")
plt.ylabel("Temperature (Â°C)")
plt.xlabel("City")
plt.xticks(rotation=45)

plt.show()
```
