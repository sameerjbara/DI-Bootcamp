```python
# --- Imports ---
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
from scipy.ndimage import rotate

# --- Load + display ---
image_path = '00_001.png'
original_image = Image.open(image_path)

plt.imshow(original_image)
plt.axis("off")
plt.show()

# --- Rotate by 30 degrees ---
def rotate_image_30_degrees(image):
    return rotate(image, 30, reshape=False, mode="nearest")

rotated = rotate_image_30_degrees(original_image)
plt.imshow(rotated)
plt.axis("off")
plt.show()

# --- Flip horizontally then vertically ---
flipped_h = ImageOps.mirror(original_image)
flipped_hv = ImageOps.flip(flipped_h)

plt.imshow(flipped_hv)
plt.axis("off")
plt.show()

# --- Zoom in (scale 1.2x) ---
w, h = original_image.size
zoomed = original_image.resize((int(w * 1.2), int(h * 1.2)))

plt.imshow(zoomed)
plt.axis("off")
plt.show()
```
